```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/Users/yuemun/Desktop/Python/Telco-Customer-Churn.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
#check for data types
df.dtypes
```

```python
#convert TotalCharges to numerical variable
df['TotalCharges'] = pd.to_numeric(df.TotalCharges, errors='coerce') 
#If error=â€˜coerceâ€™, then invalid parsing will be set as NaN.
```

```python
#check for missing values
df.isnull().sum()
#we can see there are 11 missing values in 'TotalCharges'
```

```python
#check for data with missing value in 'TotalCharges'
df[np.isnan(df.TotalCharges)]
```

```python
#we can see Tenure is zero for these rows. 
#let's check if there are any zero Tenure in other rows.
#.index-->selecting particular rows and columns of data from a DataFrame, return index number
df[df.tenure == 0].index
```

```python
#we can see there is no additional zero Tenure in other rows
```

```python
#since there are only 11 rows with zero Tenure, we can remove these rows.
#drop row(s) with label = index where tenure=0, axis=0-->drop row
df.drop(labels=df[df.tenure == 0].index, axis=0, inplace=True)
df[df['tenure'] == 0].index
```

```python
#fill in mean of TotalCharges to solve the problem of missing values in TotalCharges
values = df.TotalCharges.mean()
df.fillna(value = values)
```

```python
df.isnull().sum()
```

```python
#for SeniorCitizen column, we want to substitude 0 with 'No' and 1 with 'Yes'
df.SeniorCitizen = df.SeniorCitizen.map({0: "No", 1: "Yes"})
df.head()
```

```python
#Data Visualisation
```

```python
#1. Number of customers who churn
```

```python
ax = sns.catplot(data = df, x ="Churn", kind='count', aspect=1)
```

```python
#2. Additional services and churn 
additional_service = ["OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport", "StreamingTV", "StreamingMovies"]
#select customers who have internet service and chose to churn
df1 = df[(df.InternetService != "No") & (df.Churn == "Yes")]
#unpivot the data with additional service as variable and yes/no as value
#rename the value as 'Has service'
df1 = pd.melt(df1[additional_service]).rename({'value': 'Has service'}, axis=1)
df1.head(12)
```

```python
#plot the bar chart
ax = sns.catplot(data=df1, x='variable', hue='Has service', hue_order=['No', 'Yes'], kind='count', aspect=1.5)
ax.set(xlabel='Type of Additional Service', ylabel='Number of churns')
plt.show()
```

```python
#Customers who have Online Security, Online Backup, Device Protection, Tech Support as their additional service(s) are
# more likely to churn.
```
